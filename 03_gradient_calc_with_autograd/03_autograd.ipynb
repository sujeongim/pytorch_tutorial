```python
import torch

x = torch.ones(3, requires_grad=True) # False로 하면 Computational graph가 생기지 않음
print(x)
```

```python

y = x + 2
# computational graph
# x 
#   \
#    "+" - y - 
#   /
# 2
print(y) # grad_fn = <AddBackward0>
```

```python

z = y * y * 2 # grad_fn = <MulBackward0>
print(z)

z = z.mean() # grad_fn = <MeanBackward0>
print(z)


z.backward() # dz/dx 
# < backward >
# x가 requires_grad=True인 경우에만 가능
# z가 scalar 값인 경우에만 가능(z.mean 안하면 오류남)
print(x.grad)

# 원리 : Vec tor Jaccovian Matrix Product (J.v) -> 선대 다시 공부하기!
# eigen vector도 공부..
```

```python
z = y * y * 2 # grad_fn = <MulBackward0>
print(z)

#z = z.mean() # grad_fn = <MeanBackward0>
print(z)


z.backward() # dz/dx 
# < backward >
# x가 requires_grad=True인 경우에만 가능
# z가 scalar 값인 경우에만 가능(z.mean 안하면 오류남)
print(x.grad)

# 원리 : Vec tor Jaccovian Matrix Product (J.v) -> 선대 다시 공부하기!
# eigen vector도 공부..
```

# With no grad
(tracking 안하고 computational graph 안 그림)

Option 1 

```python
import torch

x = torch.randn(3, requires_grad=True)
print(x)

x.requires_grad_(False)
print(x)
```

Option 2

```python
import torch

x = torch.randn(3, requires_grad=True)
print(x)

x.detach()
```

Option 3

```python
import torch

x = torch.randn(3, requires_grad=True)
print(x)

with torch.no_grad():
    y = x + 2 
    print(y)
```

weights 비우기

```python
weights = torch.ones(4, requires_grad=True)

# weights.grad.zero_() 하지 않으면 축적됨
for epoch in range(2):
    model_output = (weights * 3).sum()

    model_output.backward()

    print(weights.grad)

    
```

```python
weights = torch.ones(4, requires_grad=True)

# weights.grad.zero_() 하지 않으면 축적됨
for epoch in range(2):
    model_output = (weights * 3).sum()

    model_output.backward()

    print(weights.grad)

    weights.grad.zero_()
```

optimizer

```python
weights = torch.ones(1, 4, requires_grad=True)

optimizer = torch.optim.SGD(weights, lr=0.01)
optimizer.step()
optimizer.zero_grad()
```

```python
weights = torch.ones(1, 4, requires_grad=True)

optimizer = torch.optim.SGD([weights], lr=0.01)
optimizer.step()
optimizer.zero_grad()
```
